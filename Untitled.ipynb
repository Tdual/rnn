```python
import requests
import pandas as pd

df_list = []
for i in range(1, 21):
    data = {
        "page": i,
        "per_page":100,
        "sort":"count"
    }
    r = requests.get("http://qiita.com/api/v2/tags", params=data)
    if r.status_code != 200:
        print(r.text)
    r_df = pd.read_json(r.text)
    df_list.append(r_df)

df = pd.concat(df_list)
df
```

```python
df.to_json("tags.json",orient="records")
```

```python
import requests
import urllib
def get_data(tag,page, per_page=100):
    body_url = "https://qiita.com/api/v1/tags/{}/items".format(urllib.parse.quote(tag))
    data = {
        "page": 1,
        "per_page": per_page,
        "sort":"count"
    }
    r = requests.get(body_url, data)
    if r.status_code != 200:
        print("tag:{}, \npage: {},\n message:{},".format(tag, page,r.text, ))
        return None
    return r.json()
get_data("C#",1 ,per_page=1)
```

```python
import requests
import json
import time
import urllib

uuid_set = set()

for p in range(50):
    for tag in tag_list:
        json = get_data(tag, p)
        if json:
            with table.batch_writer() as batch:
                for j in json:
                    uuid = j["uuid"]
                    if uuid not in uuid_set:
                        uuid_set.add(uuid)
                        try:
                            item={
                                 "json": j,
                                "uuid": uuid}
                            batch.put_item(item)
                        except:
                            print("same data: ", uuid)
                            pass
        time.sleep(36)
        print("finish: "+tag)
```
