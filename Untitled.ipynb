```python
import urllib
def get_data(tag,page, per_page=100):
    body_url = "https://qiita.com/api/v1/tags/{}/items".format(urllib.parse.quote(tag))
    data = {
        "page": 1,
        "per_page": per_page,
        "sort":"count"
    }
    r = requests.get(body_url, data)
    if r.status_code != 200:
        print("tag:{}, \npage: {},\n message:{},".format(tag, page,r.text, ))
        return None
    return r.json()
get_data("C#",1 ,per_page=1)
```
